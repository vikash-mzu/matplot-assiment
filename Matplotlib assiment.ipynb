```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a popular Python library used for data visualization. It allows you to create static, animated, and interactive visualizations in Python. It is widely used in data science for exploring and visualizing data, making it easier to identify trends, outliers, and patterns.

Why is it used?

It provides an easy way to plot data and customize visualizations.
It helps in generating plots, histograms, power spectra, bar charts, error charts, scatter plots, etc.
    
It s very flexible and integrates well with other libraries such as NumPy, Pandas, and SciPy.
Five types of plots that can be created using the Pyplot module of Matplotlib:

Line Plot
Bar Plot
Scatter Plot
Histogram
Pie Chart




Q2: What is a scatter plot? Use the provided code to plot a scatter plot.
A scatter plot is a type of plot used to display the relationship between two numerical variables. The values of one variable determine the position on the x-axis and the values of the other variable determine the position on the y-axis.


import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-axis (Random Data)')
plt.ylabel('Y-axis (Random Data)')

# Show plot
plt.show()


Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to divide the figure into a grid of subplots, and you can plot different graphs in each subplot, which is useful for comparing different datasets or visualizations in a single view.

Code to draw four line plots using subplot():


import numpy as np
import matplotlib.pyplot as plt

# Data for the four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)  # First subplot (2 rows, 2 columns, 1st position)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # Second subplot (2 rows, 2 columns, 2nd position)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # Third subplot (2 rows, 2 columns, 3rd position)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # Fourth subplot (2 rows, 2 columns, 4th position)
plt.plot(x, y4)
plt.title("Line 4")

# Adjust layout and show plot
plt.tight_layout()
plt.show()





Q4: What is a bar plot? Why is it used? Plot a bar plot and a horizontal bar plot using the provided data.
A bar plot is a graphical representation of categorical data where the length of each bar corresponds to the value or frequency of the category. It is used to compare different categories of data.

Code to create a vertical and horizontal bar plot:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.bar(company, profit)
plt.title('Profit by Company')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Profit by Company (Horizontal)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()




Q5: What is a box plot? Why is it used? Plot a box plot using the provided data.
A box plot (or whisker plot) is a graphical representation of the distribution of a dataset. It shows the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It is used to understand the spread, skewness, and outliers in the data.

Code to create a box plot:


import numpy as np
import matplotlib.pyplot as plt

# Data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
